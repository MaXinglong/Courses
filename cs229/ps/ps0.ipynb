# Gradients and Hessians

**(a)** Let $f(x)=\frac{1}{2}x^TAx+b^Tx$, where $A$ is symmetric and $b\in\mathbb{R}^n$. What is $\nabla f(x)$?

Solution.
$$\nabla f(x) = \nabla \left(\frac{1}{2}x^TAx+b^Tx\right) = Ax + b$$

**(b)** Let $f(x)=g(h(x))$, where $g:\mathbb{R}\rightarrow\mathbb{R}$ is differential and $h:\mathbb{R}^n\rightarrow\mathbb{R}$ is differentiable. What is $\nabla f(x)$?

Solution.
$$\nabla f(x) = \nabla g(h(x)) = g'(h(x)) \nabla h(x)$$

**(c)** Let $f(x) = \frac{1}{2}x^TAx+b^Tx$, where $A$ is symmetric and $b\in\mathbb{R}^n$ is a vector. What is $\nabla^2f(x)$?

Solution.
$$\nabla f(x) = A$$

**(d)** Let $f(x) = g(a^Tx)$, where $g:\mathbb{R}\rightarrow\mathbb{R}$ is continuously differentiable and $a\in\mathbb{R}^n$ is a vector. What are $\nabla f(x)$ and $\nabla^2 f(x)$? (Hint: your expression for $\nabla^2 f(x)$ may have as few as 11 symbols, including ' and parantheses.

Solution.
$$\nabla f(x) = \nabla g(a^Tx) = g'(a^T x)\nabla a^Tx = g'(a^Tx) a$$

Let's look at the partial derivative of one of the coordinates of $\nabla f(x)$:
$$\frac{\partial}{\partial x_j}(\nabla f(x))_i = \frac{\partial}{\partial x_j} g'(a^Tx)a_i = g''(a^Tx)a_j a_i$$
$$\Rightarrow \nabla^2 f(x) = g''(a^Tx)aa^T$$

# Positive Definite Matrices

**(a)** Let $z\in\mathbb{R}^n$ be a vector. Show that $A=zz^T$ is positive semidefinite.

Solution.
$$x^T A x = x^T z z^T x = (x^T z)^2 \geq 0$$

**(b)** Let $z\in\mathbb{R}^n$ be a non-zero vector. Let $A=z^Tz$. What is the null-space of $A$ What is the rank of $A$?

Solution.
To zero out the ith row of $A$, we can multiply the first row by $\frac{z_i}{z_1}$ and subtract that from the ith row (assuming $z_1 \neq 0$, and we can always permute the rows so that this is so since $z\neq0$). This means that the rank of $A$ is $1$, and the dimension of the null space is $n-1$. From this, we also get that

$$Ax=0 \Rightarrow z^Tx=0$$

meaning that $null(A)$ is made up of vectors orthogonal to $z$.

**(c)** Let $A\in\mathbb{R}^{n\times n}$ be a positive semidefinite matrix and $B\in\mathbb{R}^{m\times n}$ be an arbitrary matrix. Is $BAB^T$ PSD? If so, prove it. If not, give a counterexample.

Solution.
$$x^TBAB^Tx = (B^Tx)^TA(B^Tx) \geq 0$$
since $A$ is PSD.

# Eigenvectors, Eigenvalues, and the Spectral Theorem

**(a)** Suppose that the matrix $A\in\mathbb{R}^{n\times n}$ is diagonalizable, that is, $A=T\Lambda T^{-1}$ for an invertible matrix $T\in\mathbb{R}^{n\times n}$, where $\Lambda=diag(\lambda_1,\dots,\lambda_n)$ is diagonal. Denote with $t^{(i)}$ the columns of $T$. Show that $At^{(i)}=\lambda_it^{(i)}$.

Solution.
$$\begin{aligned} A  &= T\Lambda T^{-1} \\
                  AT &= T\Lambda \\
                  \Rightarrow At^{(i)} &= t^{(i)}\lambda_i
\end{aligned}$$
